```python
G = nx.Graph()


for i in atlantic['Country'].unique():
    G.add_node(i, label = i)


import networkx as nx
from pyvis.network import Network

for i in x:
    u, v = i[0], i[1]

    if G.has_edge(u, v):
        G[u][v]["weight"] += 1
    else:
        G.add_edge(u, v, weight=1)


g = Network(height="750px", width="100%", cdn_resources="in_line", notebook=True, select_menu=True, filter_menu=True)

g.add_nodes(sorted(list(atlantic['Country'].unique())), label = sorted(list(atlantic['Country'].unique())), size = sorted(list(atlantic.groupby("Country")["identifier"].count()/100)))

for u, v, data in G.edges(data=True):
    g.add_edge(u, v, value=data["weight"])

g.show_buttons(filter_=['physics'])

g.show("atlantic.html")



```

```python
g.from_nx(G) 

node_degree = {}

for Country in atlantic["Country"].unique():
    node_degree[Country] = G.degree(Country)

node_degree = dict(sorted(node_degree.items(), key=lambda x: x[1], reverse=True))

print(node_degree)

topten = list(node_degree.keys())[:10]
zeros = [node for node in node_degree.keys() if node_degree[node] == 0]

print(topten)
print(zeros)


node_degree_percentage = {}


for key in node_degree.keys():
    node_degree_percentage[key] = node_degree[key] / atlantic.groupby("Country")["identifier"].count()

#node_degree_percentage = dict(sorted(node_degree_percentage.items(), key=lambda x: x[1], reverse=True))

print(node_degree_percentage)
    

```
